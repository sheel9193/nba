# What was the average age of player in the league?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

nba = pd.read_csv('../nba/nbastats2018-2019.csv')
nba
```

```python
df = nba[['Name','Age', 'Team']]
df.describe().head(2)
```

# What player scored the most points?

```python
df1 = nba[['Name', 'Points']]
df1.groupby('Name')
points = df1.sort_values('Points', ascending = False)
points.head(1)
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
df2 = nba[['Name', 'Blocks']]
df2.groupby('Name')
blocks = df2.sort_values('Blocks', ascending = False)
blocks.head(1)
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
df3 = nba[['Name', 'Team', 'WS']]
df3.groupby('Name').max()
reg_season = df3.sort_values('WS', ascending = False)
reg_season.head(1)
```

# What player had the best 3-pt percentage? 

```python
df4 = nba[['Name','FG3%']]
df4.groupby('Name').max()
df4 = df4.sort_values('FG3%', ascending = False)
df4.head(4)
```

# Who played the most minutes during the season

```python
df5 = nba[['Name', 'MP']]
df5.groupby('Name')
df5 = df5.sort_values('MP', ascending = False)
df5.head(2)

```

# What player given their player effiecency rating was the clutchest during the season?

```python
df6 = nba[['Name', 'Team', 'Age', 'PER']]
df6 = df6.groupby('Name').max()
df6 = df6.sort_values('PER', ascending = False)
df6.head()
```

# What team had the youngest roster?

```python
df7 = nba[['Team', 'Age']]
df7 = df7.groupby('Team').mean()
df7 = df7.sort_values('Age', ascending = True)
df7.head(1).round()
```

# Who is the highest paid player during the seasion?

```python
df8 = nba [['Name', 'Team', 'Salary']]
df8 = df8.groupby('Name').max()
df8 = df8.sort_values('Salary', ascending = False)
df8.head(1)
```

# At the end of a game, who WOULDN'T you want on the Free Throw Line?

```python
df9 = nba[['Name', 'Team', 'FT%']]
df9 = df9.groupby('Name').min()
df9 = df9.sort_values('FT%')
df9.head(4)
```
